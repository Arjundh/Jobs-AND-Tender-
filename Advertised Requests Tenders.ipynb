```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support.expected_conditions import  visibility_of_element_located
import re
import pandas as pd
import numpy as np
```

```python
all_section=[]
all_link1=[]
all_link=[]
all_section1=[]
all_section2=[]
#launch url
driver = webdriver.Firefox(executable_path='C://Users//Arjun D H//geckodriver-v0.24.0-win64//geckodriver.exe')

url = "https://www.tenders.wa.gov.au/watenders/tender/search/tender-search.do;jsessionid=5F48B16B23EFBEFC67FD388F0164B245?CSRFNONCE=E0D84B4DCBF906B9EB88015375F91B3F&action=advanced-tender-search-open-tender"
driver.get(url)
driver.implicitly_wait(70)
section = driver.find_elements_by_xpath("/html[1]/body[1]/table[1]/tbody[1]/tr[2]/td[2]/table[3]/tbody[1]//td[2]")
section1 = driver.find_elements_by_xpath("/html[1]/body[1]/table[1]/tbody[1]/tr[2]/td[2]/table[3]/tbody[1]//td[3]")
section2 = driver.find_elements_by_xpath("/html[1]/body[1]/table[1]/tbody[1]/tr[2]/td[2]/table[3]/tbody[1]//td[4]")
section3 = driver.find_elements_by_xpath("/html[1]/body[1]/table[1]/tbody[1]/tr[2]/td[2]/table[3]/tbody[1]/tr/td[3]/a[1]")
for i in section:
    all_section.append(i.text)
for i in section1:
    all_section1.append(i.text)
for i in section2:
    all_section2.append(i.text)
for i in section3:
    all_link = i.get_attribute('href')
    all_link1.append(all_link)
driver.close()
```

```python
all_link1[3]
```

```python
df=pd.DataFrame(index=range(0,(len(all_section))),columns=['TenderID','TenderStatus'])
for idx,section in enumerate(all_section):
    scn=section.split('\n')
    df.iloc[idx,0]=scn[0] 
    df.iloc[idx,1]=scn[1] 
```

```python
df1=pd.DataFrame(index=range(0,(len(all_section1))),columns=['TenderTitle','PublishedVia'])
for idx,section in enumerate(all_section1):
    scn=section.split('\n')
    df1.iloc[idx,0]=scn[0] 
    df1.iloc[idx,1]=scn[1] 
```

```python
df['TenderClosingDate']=pd.Series(all_section2)
```

```python
df['TenderUrl']=pd.Series(all_link1)
```
