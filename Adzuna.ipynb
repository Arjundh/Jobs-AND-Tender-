```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support.expected_conditions import  visibility_of_element_located
from selenium.webdriver.common.action_chains import ActionChains
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
import pandas as pd
import requests
import numpy as np
import re
import os
#import javascript
#mport time
```

```python
all_section=[]
all_section1=[]
section=[]
all_link=[]
all_url=[]
all_titles=[]


#Firefox WebDriver Path
driver = webdriver.Firefox(executable_path='C://Users//Arjun D H//geckodriver-v0.24.0-win64//geckodriver.exe')

#launch url

for index in range(1,9):
    url = "https://www.adzuna.co.uk/jobs/search?loc=86383&p="+str(index)+"&pp=25&q=freelance%20Design"
    driver.get(url)
    driver.implicitly_wait(10)    
    section = driver.find_elements_by_xpath("//div[@class='a' or @class='a last']")
    for i in section:
        all_section.append(i.text)
        all_url.append(url)
    section1 = driver.find_elements_by_xpath("/html[1]/body[1]/div[1]/div[2]/div[1]/div/p[1]/span")
    for j in section1:
        all_section1.append(j.text)    
    all_title = driver.find_elements_by_xpath("/html[1]/body[1]/div[1]/div[2]/div[1]/div/h2[1]/a[1]")
    for j in all_title:
        links=j.get_attribute('href')
        all_titles.append(links)
    print(index)
driver.close()
```

```python
df=pd.DataFrame(index=range(0,len(all_section)),columns=['JobTitle','JobDetails','JobType','JobDescription','JobSalary'])
for idx,section in enumerate(all_section):
    scn=section.split('\n')
    try:
        df.iloc[idx,0]=scn[0]
    except IndexError:
        df.iloc[idx,0]=np.nan
    if('Easy apply' in scn[1]):        
        try:
            df.iloc[idx,1]=scn[2]
        except IndexError:
            df.iloc[idx,1]=np.nan
        try:
            if('Permanent' in scn[3]):
                df.iloc[idx,2]= "Permanent"
            elif('Contract' in scn[3]):
                df.iloc[idx,2]= "Contract"
            else:
                df.iloc[idx,2]=np.nan
        except IndexError:
            df.iloc[idx,2]=np.nan
        try:
            df.iloc[idx,3]=scn[3]
        except IndexError:
            df.iloc[idx,3]=np.nan
        try:
            df.iloc[idx,4]=scn[4]
        except IndexError:
            df.iloc[idx,4]=np.nan
    else:
        try:
            df.iloc[idx,1]=scn[1]
        except IndexError:
            df.iloc[idx,1]=np.nan
        try:
            if('Permanent' in scn[2]):
                df.iloc[idx,2]= "Permanent"
            elif('Contract' in scn[2]):
                df.iloc[idx,2]= "Contract"
            else:
                df.iloc[idx,2]=np.nan
        except IndexError:
            df.iloc[idx,2]=np.nan
        try:
            df.iloc[idx,3]=scn[2]
        except IndexError:
            df.iloc[idx,3]=np.nan
        try:
            df.iloc[idx,4]=scn[3]
        except IndexError:
            df.iloc[idx,4]=np.nan
```
