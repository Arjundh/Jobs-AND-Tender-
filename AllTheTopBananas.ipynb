```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support.expected_conditions import  visibility_of_element_located
import re
import pandas as pd
import os
```

```python
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait


#launch url
all_jobTitle=[]
all_jobLocation=[]
all_jobSalary=[]
all_jobOrg=[]
all_jobdesc = []
all_joburl = []

driver = webdriver.Firefox(executable_path='C://Users//Arjun D H//geckodriver-v0.24.0-win64//geckodriver.exe')

for index in range(1,16):
    url = "https://www.allthetopbananas.com/JobResults.aspx?ms=1&minsalary=0&maxsalary=0&radius=100&searchtext=design&jobtype=0&locationid=cphMainColumn_search&jobcontract=2,3&page="+str(index)
    driver.get(url)    
    print(index)
    
    driver.implicitly_wait(10)
    
    jobTitle = driver.find_elements_by_xpath("//*[@class='job-listing__title']")
    jobLocation = driver.find_elements_by_xpath("//*[@class='location']")
    jobSalary = driver.find_elements_by_xpath("//*[@class='salary']")
    jobOrg = driver.find_elements_by_xpath("//*[@class='from']")
    jobdesc = driver.find_elements_by_xpath("//*[@class='job-listing__description']")
    joburl = driver.find_elements_by_xpath("//*[@class='job-listing__title']//a")
    for j in joburl:
        all_joburl.append(j.get_attribute('href'))
    for tt in jobdesc:
        all_jobdesc.append(tt.text)
    for i in jobOrg:
        all_jobOrg.append(i.text)
    for m in jobSalary:
        all_jobSalary.append(m.text)
    for n in jobLocation:
        all_jobLocation.append(n.text)
    for p in jobTitle:
        all_jobTitle.append(p.text)
driver.close()

```

```python
print(len(all_joburl))
```

```python
df=pd.DataFrame(index=range(0,(len(all_jobTitle))),columns=['JobTitle','JobCompanyName','JobLocation','JobSalary','JobDescription','JobUrl'])
```

```python
df['JobTitle'] = pd.Series(all_jobTitle)
df['JobCompanyName'] = pd.Series(all_jobOrg)
df['JobLocation'] = pd.Series(all_jobLocation)
df['JobSalary'] = pd.Series(all_jobSalary)
df['JobDescription'] = pd.Series(all_jobdesc)
df['JobUrl'] = pd.Series(all_joburl)
```
