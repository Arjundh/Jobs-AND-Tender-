```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support.expected_conditions import  visibility_of_element_located
import re
import pandas as pd
import numpy as np
```

```python
all_section=[]
all_section1=[]
all_section2=[]
all_section3=[]
all_link1=[]
all_link=[]
#launch url
driver = webdriver.Firefox(executable_path='C://Users//Arjun D H//geckodriver-v0.24.0-win64//geckodriver.exe')

for index in range(10,25):
    print(index)
    url = "https://aquent.com/find-work/?page="+str(index)+"&l=all&k=&ux=on&design=on&p3=on&p2=on&p1=on&site1=on"
    driver.get(url)
    driver.implicitly_wait(70)
    section1 = driver.find_elements_by_xpath("//h2[@class='sans-serif fixed-size-large text-capitalize']//a")
    section = driver.find_elements_by_xpath("//span[@class='location margin-bottom-1']")
    #section2 = driver.find_elements_by_xpath("//span[@class='jobposting-location']")
    #section3 = driver.find_elements_by_xpath("//p[@class='jobposting-snippet']")
    for i in section1:
        all_link = i.get_attribute('href')
        all_link1.append(all_link)
        all_section.append(i.text)  
    for i in section:
        all_section1.append(i.text)
    #for i n section2:
        #all_section2.append(i.text)
   # for i in section3:
       # all_section3.append(i.text)  
driver.close()
```

```python
all_section8=[]
all_section9=[]
driver = webdriver.Firefox(executable_path='C://Users//Arjun D H//geckodriver-v0.24.0-win64//geckodriver.exe')

for item in all_link1:
    driver.get(item)
    driver.implicitly_wait(10)
    section = driver.find_elements_by_xpath("//div[@class='job-info']")
    section1 = driver.find_elements_by_xpath("//div[@class='job-desc job-details-section']")
    for i in section:
        all_section8.append(i.text)
    for i in section1:
        all_section9.append(i.text)
           
```

```python
df=pd.DataFrame(index=range(0,(len(all_section))),columns=['JobTitle'])
for idx,section in enumerate(all_section):
    scn=section.split('\n')
    df.iloc[idx,0]=scn[0] 
    
    
```

```python
df1=pd.DataFrame(index=range(0,(len(all_section8))),columns=['JobLocation','JobSalary'])
for idx,section in enumerate(all_section8):
    scn=section.split('\n')
    df1.iloc[idx,0]=scn[0] 
    #df1.iloc[idx,1]=scn[1] 
    #df1.iloc[idx,2]=scn[2] 
    #df1.iloc[idx,3]=scn[3]
    try:
        df1.iloc[idx,0]=scn[0]
    except IndexError:
        df1.iloc[idx,0]=np.nan 
    if('Salary:' in scn[2]):     
        try:
            s1=scn[2].split('Salary:')
            df1.iloc[idx,1]=s1[1]
        except IndexError:
            df1.iloc[idx,1]=np.nan 
    else:
        try:
            df1.iloc[idx,1]=np.nan #job title
        except IndexError:
            df1.iloc[idx,1]=np.nan    
        
    
```
