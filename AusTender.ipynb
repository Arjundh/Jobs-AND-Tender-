```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support.expected_conditions import  visibility_of_element_located
import re
import pandas as pd
import numpy as np
```

```python
all_section=[]
all_section1=[]
all_section2=[]
all_section3=[]
all_link1=[]
all_link=[]
#launch url
driver = webdriver.Firefox(executable_path='C://Users//Arjun D H//geckodriver-v0.24.0-win64//geckodriver.exe')

for index in range(1,30):
    print(index)
    url = "https://www.tenders.gov.au/Cn/List?Weekly=01-Sep-2019%2c07-Sep-2019&Weekly_Submit=&page="+str(index)
    driver.get(url)
    driver.implicitly_wait(70)
    section1 = driver.find_elements_by_xpath("//p[@class='lead']")
    section = driver.find_elements_by_xpath("//div[@class='list-desc-inner']//a")
    section2 = driver.find_elements_by_xpath("//div[@class='box boxW listInner']")
    section3 = driver.find_elements_by_xpath("//p[@class='lead']")
    for i in section1:
        all_section.append(i.text)  
    for i in section:
        all_section1.append(i.text)
        all_link = i.get_attribute('href')
        all_link1.append(all_link)
    for i in section2:
        all_section2.append(i.text)
    for i in section3:
        all_section3.append(i.text)  
driver.close()
```

```python
df=pd.DataFrame(index=range(0,(len(all_section2))),columns=['TenderID','PublishedVia','TenderPublishDate','TenderContractPeriod','TenderValue','Supplier Name','Last Updated'])
for idx,section in enumerate(all_section2):
    scn=section.split('\n')
    df.iloc[idx,0]=scn[1]
    df.iloc[idx,1]=scn[3]
    df.iloc[idx,2]=scn[5]
    df.iloc[idx,3]=scn[7]
    df.iloc[idx,4]=scn[9]
    df.iloc[idx,5]=scn[11]
    df.iloc[idx,6]=scn[13]
```
