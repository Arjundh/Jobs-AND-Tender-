```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support.expected_conditions import  visibility_of_element_located
import re
import pandas as pd
import numpy as np
```

```python
all_section = []
all_section1 = []
all_section2 = []
all_section3 = []
all_section4 = []
all_section5 = []
all_section6 = []
all_section7 =[]
all_link1=[]
driver = webdriver.Firefox(executable_path='C://Users//Arjun D H//geckodriver-v0.24.0-win64//geckodriver.exe')
list1 = ['Graphic Design']
         

for item in list1:
    category1 = item.split('\n')

    for index in range(0,4):
        url = "https://www.britishjobs.co.uk/search-jobs?category=&distance=15&fp=1&geo=&offset="+str(index)+"00&perpage=100&posted=&q="+item+"&salarymax=&salarymin=&salarytype=annum&search=1&tempperm=Any"
        driver.get(url)
        print(index)
    
        driver.implicitly_wait(10)
    
        section = driver.find_elements_by_xpath("//h2[@class='job__title']//a")
        section1 = driver.find_elements_by_xpath("//dd[@class='job__details-value location']")
        section2 = driver.find_elements_by_xpath("//p[@class='job__description noscript-show']")
        section3 = driver.find_elements_by_xpath("//dd[@class='job__details-value salary']")
        section4 = driver.find_elements_by_xpath("//dl[@class='job__details']")
        section5 = driver.find_elements_by_xpath("//dl[@class='job__details l-flex-row']")
        section6 = driver.find_elements_by_xpath("//dd[@class='job__details-value text--semibold']")
        for i in section:
            all_section.append(i.text)
            all_link = i.get_attribute('href')
            all_link1.append(all_link)
        for i in section1:
            all_section1.append(i.text)
            all_section6.append(category1)
        for i in section2:
            all_section2.append(i.text)
        for i in section3:
            all_section3.append(i.text)
        for n in section4:
            all_section4.append(n.text)
        for q in section5:
            all_section5.append(q.text)
        for q in section6:
            all_section7.append(q.text)
driver.close()
```

```python
df=pd.DataFrame(index=range(0,(len(all_section5))),columns=['JobSalary'])
for idx,section in enumerate(all_section5):
    scn=section.split('\n')
    if('£' in scn[1]):     
        try:
            df.iloc[idx,0]=scn[1]
        except IndexError:
            df.iloc[idx,0]=np.nan 
    
```
