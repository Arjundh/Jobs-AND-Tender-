```python
#launch url
all_titles=[]
all_description=[]
all_url=[]
all_section=[]
all_section1 = []
all_link = []
all_link1 = []
all_section2 = []
all_section3 = []
all_section4 = []
all_section5 = []

driver = webdriver.Firefox(executable_path='C://Users//Arjun D H//geckodriver-v0.24.0-win64//geckodriver.exe')


url = "https://www.careerjunction.co.za/jobs/results?keywords=design&jobtypeid=27&jobtypeid=25&jobtypeid=31&categoryid=5#results"
driver.get(url)
driver.implicitly_wait(30)
section = driver.find_elements_by_xpath("//a[@class='noUnderline jobTitle']")
section1 = driver.find_elements_by_xpath("//span[@itemprop='employmentType']")
section2 = driver.find_elements_by_xpath("//span[@itemprop='jobLocation']")
section3 = driver.find_elements_by_xpath("//span[@itemprop='baseSalary']")
section4 = driver.find_elements_by_xpath("//span[@itemprop='datePosted']")
section5 = driver.find_elements_by_xpath("//span[@itemprop='hiringOrganization']")

for i in section:
    all_section.append(i.text) 
for i in section1:
    all_section1.append(i.text)
for i in section2:
    all_section2.append(i.text)
for i in section3:
    all_section3.append(i.text)
for i in section4:
    all_section4.append(i.text)
for i in section5:
    all_section5.append(i.text)
    all_title = driver.find_elements_by_xpath("//a[@class='noUnderline jobTitle']")
for i in section:
    all_link = i.get_attribute('href')
    all_link1.append(all_link)
driver.close()
```

```python
df=pd.DataFrame(index=range(0,len(all_section)),columns=['JobTitle'])
for idx,section in enumerate(all_section):
    scn=section.split('\n')
    try:
        df.iloc[idx,0]=scn[0]
    except IndexError:
        df.iloc[idx,0]=np.nan
    
```

```python
df['JobCompanyName']=pd.Series(all_section5)
df['JobLocation']=pd.Series(all_section2)
#df['Job Type']=pd.Series(all_section1)
df['JobSalary']=pd.Series(all_section3)
#df['JobPosted Date']=pd.Series(all_section)
df['JobUrl'] = pd.Series(all_link1)
df['LogoUrl'] = "https://www.stepstone.com/wp-content/themes/StSt-Com-v1/images/timeline/logo_timeline_careerjunction_2.jpg"
```
