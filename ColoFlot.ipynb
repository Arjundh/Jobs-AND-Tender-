```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support.expected_conditions import  visibility_of_element_located
from selenium.webdriver.common.action_chains import ActionChains
import re
import pandas as pd
import numpy as np
```

```python
#launch url
all_titles=[]
all_url=[]
all_section=[]
all_section1=[]
all_section2=[]
all_location = []
all_link1 = []
driver = webdriver.Firefox(executable_path='C://Users//Arjun D H//geckodriver-v0.24.0-win64//geckodriver.exe')

url = "https://www.coroflot.com/design-jobs/remote"
driver.get(url)
driver.implicitly_wait(10)
    
jobsection = driver.find_elements_by_xpath("/html[1]/body[1]/div[1]/div[3]/section[2]/section[1]/div[1]/div[1]/ul[1]/li/a[1]")
section = driver.find_elements_by_xpath("/html[1]/body[1]/div[1]/div[3]/section[2]/section[1]/div[1]/div[1]/ul[1]/li/a[1]/span[1]/strong[1]")
section1 = driver.find_elements_by_xpath("/html[1]/body[1]/div[1]/div[3]/section[2]/section[1]/div[1]/div[1]/ul[1]/li/a[1]/span[2]")
for j in jobsection:
    all_section.append(j.text)
    all_url.append(j.get_attribute('href'))
for tt in section:
    all_section1.append(tt.text)
for t in section1:
    all_section2.append(t.text)    
driver.close()
```

```python
df=pd.DataFrame(index=range(0,len(all_section)),columns=['JobTitle','JobCompanyName'])
for idx,section in enumerate(all_section):
    scn=section.split('\n')
    try:
        df.iloc[idx,0]=scn[0]
    except IndexError:
        df.iloc[idx,0]=np.nan
    try:
        if(len(scn[1])>1):
            df.iloc[idx,1]=scn[1]
            
        else:
            df.iloc[idx,1]=scn[2]
            
    except IndexError:
        df.iloc[idx,1]=np.nan
    
    
```

```python
df['JobUrl']=pd.Series(all_url)
df['JobLocation']=pd.Series(all_section2)
df['LogoUrl']="https://mir-s3-cdn-cf.behance.net/project_modules/max_1200/7a046c33590213.56b0ca1517574.png"
```
