```python
#launch url
all_titles=[]
all_category=[]
all_url=[]
all_section=[]
all_section1 = []

driver = webdriver.Firefox(executable_path='C://Users//Arjun D H//geckodriver-v0.24.0-win64//geckodriver.exe')
for index in range(1,15):    
    url = "https://jobs.designcrowd.com/jobs/open/"+str(index)+"/"
    driver.get(url)
    driver.implicitly_wait(10)
    section = driver.find_elements_by_xpath("//div[@class='project-card__main']")
    for i in section:
        all_section.append(i.text)
        all_url.append(url)
    all_title = driver.find_elements_by_xpath("/html[1]/body[1]/div[1]/main[1]/section[2]/div[1]/div[3]/div/div[1]/div[1]/div[2]/div[1]/div[1]/div[1]/div[1]/h4[1]/a[1]")
    for j in all_title:
        links=j.get_attribute('href')
        all_titles.append(links)
    print(index)
driver.close()
```
