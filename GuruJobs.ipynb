```python
#launch url
all_section=[]
all_section1=[]
all_section2=[]
all_section3=[]
all_section4=[]
#all_section5=[]
all_link1=[]
all_link=[]

driver = webdriver.Firefox(executable_path='C://Users//Arjun D H//geckodriver-v0.24.0-win64//geckodriver.exe')

for index in range(1,20):    
    url = "https://www.guru.com/d/jobs/c/design-art/pg/"+str(index)+"/"
    driver.get(url)
    driver.implicitly_wait(70)
    section = driver.find_elements_by_xpath("/html[1]/body[1]/form[1]/main[1]/main[1]/section[1]/div[1]/div[2]/div[1]/ul[1]/li/div[1]/div[2]/div[2]/div[1]/p[1]")
    section1 = driver.find_elements_by_xpath("/html[1]/body[1]/form[1]/main[1]/main[1]/section[1]/div[1]/div[2]/div[1]/ul[1]/li/div[1]/div[2]/div[1]/div[1]/h2[1]")
    section2 = driver.find_elements_by_xpath("/html[1]/body[1]/form[1]/main[1]/main[1]/section[1]/div[1]/div[2]/div[1]/ul[1]/li/div[1]/div[2]/div[2]/div[1]/div[1]")
    section3 = driver.find_elements_by_xpath("/html[1]/body[1]/form[1]/main[1]/main[1]/section[1]/div[1]/div[2]/div[1]/ul[1]/li/div[1]/div[2]/div[2]/div[2]/div[2]/p[1]/strong[1]")
    section4 = driver.find_elements_by_xpath("/html[1]/body[1]/form[1]/main[1]/main[1]/section[1]/div[1]/div[2]/div[1]/ul[1]/li/div[1]/div[2]/div[1]/div[1]/h2[1]/a[1]")
    #section5 = driver.find_elements_by_xpath("/html[1]/body[1]/form[1]/main[1]/main[1]/section[1]/div[1]/div[2]/div[1]/ul[1]/li/div[1]/div[2]/div[1]/div[1]/div[2]/strong[2]")
    
    for i in section:
        
        all_section.append(i.text)
    for i in section1:
        all_section1.append(i.text)
    for i in section2:
        all_section2.append(i.text)
    for i in section3:
        all_section3.append(i.text)
    #for i in section5:
        #all_section5.append(i.text)    
    for i in section4:
        all_link = i.get_attribute('href')
        all_link1.append(all_link)
```
