```python
import pandas as pd
import numpy as np

from dateutil.parser import parse
```

```python
all_section=[]
all_section1=[]
all_section2=[]
all_section3=[]
all_section4=[]
all_link1=[]
all_link=[]
#launch url
driver = webdriver.Firefox(executable_path='C://Users//Arjun D H//geckodriver-v0.24.0-win64//geckodriver.exe')
for index in range(0,8):
    print(index)
    url = "https://www.isro.gov.in/tenders?field_advertiser_value=All&field_tender_date_value_1[value]=&field_tender_date_value2_1[value]=&page="+str(index)
    driver.get(url)
    driver.implicitly_wait(70)
    section = driver.find_elements_by_xpath("/html[1]/body[1]/div[1]/div[4]/div[2]/div[1]/div[1]/div[1]/div[1]/div[1]/div[1]/div[2]/div[1]/div[1]/div[1]/div[1]/div[2]/table[1]/tbody[1]/tr/td[2]")
    section1 = driver.find_elements_by_xpath("/html[1]/body[1]/div[1]/div[4]/div[2]/div[1]/div[1]/div[1]/div[1]/div[1]/div[1]/div[2]/div[1]/div[1]/div[1]/div[1]/div[2]/table[1]/tbody[1]/tr/td[3]")
    section2 = driver.find_elements_by_xpath("/html[1]/body[1]/div[1]/div[4]/div[2]/div[1]/div[1]/div[1]/div[1]/div[1]/div[1]/div[2]/div[1]/div[1]/div[1]/div[1]/div[2]/table[1]/tbody[1]/tr/td[4]")
    section3 = driver.find_elements_by_xpath("/html[1]/body[1]/div[1]/div[4]/div[2]/div[1]/div[1]/div[1]/div[1]/div[1]/div[1]/div[2]/div[1]/div[1]/div[1]/div[1]/div[2]/table[1]/tbody[1]/tr/td[5]")
    section4 = driver.find_elements_by_xpath("//td[@class='views-field views-field-title']//a")
    for i in section:
        
        all_section.append(i.text)
    for i in section1:
        all_section1.append(i.text)
    for i in section2:
        all_section2.append(i.text)
    for i in section3:
        all_section3.append(i.text)    
    for i in section4:
        all_link = i.get_attribute('href')
        all_link1.append(all_link)
```

```python
df=pd.DataFrame(index=range(0,(len(all_section))),columns=['TenderTitle'])
for idx,section in enumerate(all_section):
    scn=section.split('\n')
    df.iloc[idx,0]=scn[0] 
```
