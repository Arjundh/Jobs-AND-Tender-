```python
#launch url
all_titles=[]
all_category=[]
all_url=[]
all_section=[]
all_section1 = []

driver = webdriver.Firefox(executable_path='C://Users//Arjun D H//geckodriver-v0.24.0-win64//geckodriver.exe')

#for index in range(0,45):    
url = "https://jobspresso.co/remote-work/#s=1"
driver.get(url)
driver.implicitly_wait(10)
time.sleep(10)
driver.find_element_by_xpath("//input[@id='search_keywords']").clear()
#driver.implicitly_wait(10)
#driver.find_element_by_xpath("//input[id='search_keywords']").send_keys("Designer")
time.sleep(1)
driver.find_element_by_xpath("//label[@class='developer']").click()
time.sleep(1)
driver.find_element_by_xpath("//label[@class='devops']").click()
#driver.implicitly_wait(10)
time.sleep(1)
driver.find_element_by_xpath("//label[@class='marketing']").click()
#driver.implicitly_wait(10)
time.sleep(1)
driver.find_element_by_xpath("//label[@class='pm']").click()
#driver.implicitly_wait(10)
time.sleep(1)
driver.find_element_by_xpath("//label[@class='sales']").click()
#driver.implicitly_wait(10)
time.sleep(1)
driver.find_element_by_xpath("//label[@class='support']").click()
time.sleep(2)

driver.find_element_by_xpath("//input[@name='submit']").click()
driver.implicitly_wait(10)
for index in range(0,2):
    driver.implicitly_wait(10)
    driver.find_element_by_xpath("//a[@class='load_more_jobs']").click()
    driver.implicitly_wait(10)
    time.sleep(10)
    print(index)
driver.execute_script("scrollBy(0,250);")
time.sleep(10)
section = driver.find_elements_by_xpath("//div[@class='job_listing-about']")
for i in section:
    all_section.append(i.text)
    all_url.append(url)
all_title = driver.find_elements_by_xpath("//a[@class='job_listing-clickbox']")
for j in all_title:
    links=j.get_attribute('href')
    all_titles.append(links)
    
    
driver.close()
```

```python
df=pd.DataFrame(index=range(0,len(all_section)),columns=['JobTitle','JobDescription','JobLocation'])
for idx,section in enumerate(all_section):
    scn=section.split('\n')
    try:
        df.iloc[idx,0]=scn[0]
    except IndexError:
        df.iloc[idx,0]=np.nan
    try:
        df.iloc[idx,1]=scn[1]
    except IndexError:
        df.iloc[idx,1]=np.nan
    try:
        df.iloc[idx,2]=scn[2]
    except IndexError:
        df.iloc[idx,2]=np.nan
    
```

```python
df['JobUrl']=pd.Series(all_titles)
df['LogoUrl']='https://finalmonsoon.files.wordpress.com/2019/09/jobspresso-logo-400x118.jpg?w=840'
```
