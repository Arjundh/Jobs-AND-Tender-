```python
all_section=[]
#all_section1=[]
#all_section2=[]
all_section3=[]
#all_section4=[]
all_section5=[]
all_link1=[]
all_link=[]
#launch url
driver = webdriver.Firefox(executable_path='C://Users//Arjun D H//geckodriver-v0.24.0-win64//geckodriver.exe')
for index in range(1,104):
    print(index)
    url = "https://procontract.due-north.com/Opportunities/Index?Page="+str(index)+"&SortColumn=Title&SortDirection=Ascending"
    driver.get(url)
    driver.implicitly_wait(70)
    section = driver.find_elements_by_xpath("/html[1]/body[1]/div[1]/main[1]/form[1]/div[4]/div[1]/div[1]/div[1]/div[1]/div[2]/table[1]/tbody[1]/tr/td[1]/a[1]")
    #section1 = driver.find_elements_by_xpath("/html[1]/body[1]/div[1]/div[1]/main[1]/form[1]/div[4]/div[2]/div[1]/div[1]/div[1]/div[2]/table[1]/tbody[1]/tr/td[2]")
    #section2 = driver.find_elements_by_xpath("/html[1]/body[1]/div[1]/div[1]/main[1]/form[1]/div[4]/div[2]/div[1]/div[1]/div[1]/div[2]/table[1]/tbody[1]/tr/td[3]")
    section3 = driver.find_elements_by_xpath("/html[1]/body[1]/div[1]/main[1]/form[1]/div[4]/div[1]/div[1]/div[1]/div[1]/div[2]/table[1]/tbody[1]/tr/td[4]")
    #section4 = driver.find_elements_by_xpath("/html[1]/body[1]/div[1]/div[1]/main[1]/form[1]/div[4]/div[2]/div[1]/div[1]/div[1]/div[2]/table[1]/tbody[1]/tr/td[5]")
    section5 = driver.find_elements_by_xpath("/html[1]/body[1]/div[1]/main[1]/form[1]/div[4]/div[1]/div[1]/div[1]/div[1]/div[2]/table[1]/tbody[1]/tr/td[1]/a[1]")
    for i in section:
        
        all_section.append(i.text)
    #for i in section1:
        #all_section1.append(i.text)
    #for i in section2:
        #all_section2.append(i.text)
    for i in section3:
        all_section3.append(i.text)    
    #for i in section4:
        #all_section4.append(i.text)       
    for i in section5:
        all_link = i.get_attribute('href')
        all_link1.append(all_link)
```
