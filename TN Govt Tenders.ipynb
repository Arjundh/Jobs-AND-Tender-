```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support.expected_conditions import  visibility_of_element_located
import re
import pandas as pd
import numpy as np
```

```python
all_section=[]
all_section1=[]
all_section2=[]
all_section3=[]
all_section4=[]
all_section5=[]
all_link1=[]
all_link=[]
#launch url
driver = webdriver.Firefox(executable_path='C://Users//Arjun D H//geckodriver-v0.24.0-win64//geckodriver.exe')
for index in range(1,15):
    print(index)
    url = "http://www.tenders.tn.gov.in/innerpage.asp?choice=ot4&page="+str(index)
    driver.get(url)
    driver.implicitly_wait(70)
    section = driver.find_elements_by_xpath("/html[1]/body[1]/table[1]/tbody[1]/tr[4]/td[2]/table[1]/tbody[1]/tr[1]/td[2]/table[1]/tbody[1]/tr[1]/td[1]/table[1]/tbody[1]/tr[4]/td[1]/table[1]/tbody[1]//td[1]")
    section1 = driver.find_elements_by_xpath("/html[1]/body[1]/table[1]/tbody[1]/tr[4]/td[2]/table[1]/tbody[1]/tr[1]/td[2]/table[1]/tbody[1]/tr[1]/td[1]/table[1]/tbody[1]/tr[4]/td[1]/table[1]/tbody[1]//td[2]")
    section2 = driver.find_elements_by_xpath("/html[1]/body[1]/table[1]/tbody[1]/tr[4]/td[2]/table[1]/tbody[1]/tr[1]/td[2]/table[1]/tbody[1]/tr[1]/td[1]/table[1]/tbody[1]/tr[4]/td[1]/table[1]/tbody[1]//td[3]")
    section3 = driver.find_elements_by_xpath("/html[1]/body[1]/table[1]/tbody[1]/tr[4]/td[2]/table[1]/tbody[1]/tr[1]/td[2]/table[1]/tbody[1]/tr[1]/td[1]/table[1]/tbody[1]/tr[4]/td[1]/table[1]/tbody[1]//td[4]")
    section4 = driver.find_elements_by_xpath("/html[1]/body[1]/table[1]/tbody[1]/tr[4]/td[2]/table[1]/tbody[1]/tr[1]/td[2]/table[1]/tbody[1]/tr[1]/td[1]/table[1]/tbody[1]/tr[4]/td[1]/table[1]/tbody[1]/tr/td[5]")
    section5 = driver.find_elements_by_xpath("//td[@width='40%']//a")
    for i in section:
        
        all_section3.append(i.text)
    for i in section1:
        all_section1.append(i.text)
    for i in section2:
        all_section2.append(i.text)
    for i in section3:
        all_section.append(i.text)    
    for i in section4:
        all_section4.append(i.text)       
    for i in section5:
        all_link = i.get_attribute('href')
        all_link1.append(all_link)
```
