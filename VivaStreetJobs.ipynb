```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support.expected_conditions import  visibility_of_element_located
from selenium.webdriver.common.action_chains import ActionChains
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
import pandas as pd
import requests
import numpy as np
import re
import os
#import javascript
#mport time
```

```python
#launch url
all_titles=[]
all_category=[]
all_url=[]
all_section=[]
all_section1 = []

driver = webdriver.Firefox(executable_path='C://Users//Arjun D H//geckodriver-v0.24.0-win64//geckodriver.exe')

for index in range(1,4):    
    url = "https://search.vivastreet.co.uk/job-vacancies/gb/t+"+str(index)+"?lb=new&search=1&start_field=1&keywords=designer&cat_1=1&geosearch_text=&searchGeoId=&sp_communities_rideshare_detail=Contract%2FFreelance"
    driver.get(url)
    driver.implicitly_wait(10)
    section = driver.find_elements_by_xpath("//div[@class='clad__summary']")
    for i in section:
        all_section.append(i.text)
        all_url.append(url)
    all_title = driver.find_elements_by_xpath("//a[@class='clad__wrapper']")
    for j in all_title:
        links=j.get_attribute('href')
        all_titles.append(links)
    print(index)
driver.close()
```

```python
print(len(all_titles))
```

```python
df=pd.DataFrame(index=range(0,len(all_section)),columns=['JobTitle','JobType','JobLocation','JobDescription'])
for idx,section in enumerate(all_section):
    scn=section.split('\n')
    try:
        df.iloc[idx,0]=scn[0]
    except IndexError:
        df.iloc[idx,0]=np.nan
    try:        
        s1=scn[1].split(' ')
        df.iloc[idx,1]=s1[0]     
        newlst = s1[1:len(s1)]
        skills = ' '.join(newlst)
        df.iloc[idx,2]=skills
    except IndexError:
        df.iloc[idx,2]=np.nan
        df.iloc[idx,1]=np.nan
    try:
        df.iloc[idx,3]=scn[2]
    except IndexError:
        df.iloc[idx,3]=np.nan
```

```python
df['JobUrl']=pd.Series(all_titles)
df['LogoUrl']="https://s3-eu-west-1.amazonaws.com/tpd/logos/4fedd6e600006400051737ab/0x0.png"
df["isForFreeUsers"]="No"
#df['Job Description']=pd.Series(all_section1)
```
