```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support.expected_conditions import  visibility_of_element_located
from selenium.webdriver.common.action_chains import ActionChains
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
import pandas as pd
import requests
import numpy as np
import re
import os
#import javascript
#mport time
```

```python
#launch url
all_titles=[]
all_category=[]
all_url=[]
all_section=[]
all_section1 = []

driver = webdriver.Firefox(executable_path='C://Users//Arjun D H//geckodriver-v0.24.0-win64//geckodriver.exe')

#for index in range(0,45):    
url = "https://www.workingnomads.co/jobs?category=design"
driver.get(url)
driver.implicitly_wait(10)
section = driver.find_elements_by_xpath("//div[@class='job clearfix post' or @class='job clearfix post  highlighted_true' or @class='job clearfix post  expired']")
for i in section:
    all_section.append(i.text)
    all_url.append(url)
all_title = driver.find_elements_by_xpath("//a[@class='open-button  ng-binding']")
for j in all_title:
    links=j.get_attribute('href')
    all_titles.append(links)
    #print(index)
driver.close()
```

```python
df=pd.DataFrame(index=range(0,len(all_section)),columns=['JobTitle','JobCompanyName','JobMainCategory','JobType','JobSkills'])
for idx,section in enumerate(all_section):
    scn=section.split('\n')
    try:
        df.iloc[idx,0]=scn[2]
    except IndexError:
        df.iloc[idx,0]=np.nan
    try:
        df.iloc[idx,1]=scn[0]
    except IndexError:
        df.iloc[idx,1]=np.nan
    try:        
        s1=scn[3].split(' ')
        df.iloc[idx,2]=s1[0]
        df.iloc[idx,3]=s1[1]        
        newlst = s1[2:len(s1)]
        skills = ' '.join(newlst)
        df.iloc[idx,4]=skills
    except IndexError:
        df.iloc[idx,2]=np.nan
    
```

```python
df['JobUrl'] = pd.Series(all_titles)
df['LogoUrl'] = "https://finalmonsoon.files.wordpress.com/2018/03/working-nomads-big-logo-1.jpg?w=900"
```
