```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support.expected_conditions import  visibility_of_element_located
import re
import pandas as pd
import numpy as np
```

```python
all_section=[]
all_link1=[]
all_link=[]
all_section2=[]
#launch url
driver = webdriver.Firefox(executable_path='C://Users//Arjun D H//geckodriver-v0.24.0-win64//geckodriver.exe')

for index in range(0,3):
    print(index)
    url = "https://www.awwwards.com/jobs/search/?filter_by_type=offer&category=design&page="+str(index)+""
    driver.get(url)
    driver.implicitly_wait(70)
    section1 = driver.find_elements_by_xpath("//*[@class='content']")
    section = driver.find_elements_by_xpath("//a[@class='profile-link']")
    #section2 = driver.find_elements_by_xpath("//div[@class='box-right']")
    for i in section1:
        all_section.append(i.text)
    for i in section:
        all_link = i.get_attribute('href')
        #print(all_link)
        all_link1.append(all_link)
    #for i in section2:
        #all_section2.append(i.text)
```

```python
df=pd.DataFrame(index=range(0,(len(all_section))),columns=['JobLocation','JobTitle','JobCompanyName','JobDescription'])
for idx,section in enumerate(all_section):
    scn=section.split('\n')
    df.iloc[idx,0]=scn[0] 
    df.iloc[idx,1]=scn[1]
    df.iloc[idx,2]=scn[2]
    df.iloc[idx,3]=scn[3]

            
    
```
