```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support.expected_conditions import  visibility_of_element_located
import re
import pandas as pd
import numpy as np
```

```python
all_section=[]
all_section1=[]
all_section2=[]
all_section3=[]
all_section4=[]
all_section5=[]
all_link1=[]
all_link=[]
#launch url
driver = webdriver.Firefox(executable_path='C://Users//Arjun D H//geckodriver-v0.24.0-win64//geckodriver.exe')
for index in range(0,50):
    print(index)
    url = "https://www.contractsfinder.service.gov.uk/Search/Results?&page="+str(index)
    driver.get(url)
    driver.implicitly_wait(70)
    section1 = driver.find_elements_by_xpath("//div[@class='search-result-header']//a")
    #section = driver.find_elements_by_xpath("//div[@class='wrap-text']")
    #section2 = driver.find_elements_by_xpath("//div[@class='search-result-sub-header wrap-text']")
    #section3 = driver.find_elements_by_xpath("//div[@class='search-result-entry']")
    section4 = driver.find_elements_by_xpath("//div[@class='search-result']")
    #section5 = driver.find_elements_by_xpath("//div[@class='search-result-entry']")
    for i in section1:
        all_link = i.get_attribute('href')
        all_link1.append(all_link)
    #for i in section:
        #all_section1.append(i.text)
    #for i in section2:
        #all_section2.append(i.text)
    #for i in section3:
        #all_section3.append(i.text)    
    for i in section4:
        all_section4.append(i.text)    
    #for i in section5:
        #all_section5.append(i.text)    
```

```python
all_section4[0]
```

```python
df=pd.DataFrame(index=range(0,(len(all_section4))),columns=['TenderTitle','PublishedVia','TenderLocation','TenderClosingDate'])
for idx, myl in enumerate(all_section4):
    mul=myl.split('\n')
    try:
        df.iloc[idx,0]=mul[0]
    except IndexError:
        df.iloc[idx,1]=mul[1]
    try:
        df.iloc[idx,1]=mul[1]
    except IndexError:
        df.iloc[idx,1]=np.nan
    
    try:
        df.iloc[idx,2]=mul[3]
    except IndexError:
        df.iloc[idx,2]=np.nan
   
    
```
